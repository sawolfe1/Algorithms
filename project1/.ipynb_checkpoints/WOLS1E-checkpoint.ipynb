```python
from matplotlib import pyplot as plt
import timeit

# recursive
rFibS = """
def rFib(num): 
    sum = 0
    if num < 2:
        return num
    else:
        sum = rFib(num - 2) + rFib(num - 1)
    return sum
"""

fibNumber1 = []
executionTime1 = []
for i in range(30):
    fibNumber1.append(i)
    executionTime1.append(timeit.timeit(
        stmt=f"rFib({i})", setup=rFibS, number=10))

plt.plot(fibNumber1, executionTime1)
plt.xlabel("N")
plt.ylabel("Time (S)")
```

```python
# iterative
fibS = """
def fib(num):  # non-recursive
    f = [0, 1]
    for i in range(1, num):
        sum = f[i - 1] + f[i]
        f.append(sum)
    return f[num]
"""

fibNumber2 = []
executionTime2 = []
for i in range(30):
    fibNumber2.append(i)
    executionTime2.append(timeit.timeit(
        stmt=f"fib({i})", setup=fibS, number=10))

plt.plot(fibNumber2, executionTime2, label="Iterative")
plt.xlabel("N")
plt.ylabel("Time(s)")
```

```python
plt.plot(fibNumber1, executionTime1, label="Recursive")
plt.plot(fibNumber2, executionTime2, label="Iterative")

plt.xlabel("N")
plt.ylabel("Time (S)")
plt.legend()
```
